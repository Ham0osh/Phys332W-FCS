```
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```
def f_mass(vol,rho):
    return rho*vol
```

```
vol, mass = np.genfromtxt('/work/data/micro_pipette_calibration.txt', skip_header=2, unpack=True, delimiter=',')
fig, (ax1,ax2) = plt.subplots(2,1, figsize = (8,3*2), gridspec_kw = {'height_ratios': (2,1.2), 'hspace' : 0.4})

# plt.subplots_adjust(hspace=0.4)

ax1.plot(vol,mass*1000,'.')

pOpt, pCov = curve_fit(f_mass, vol, mass, p0=(0.001))
rho = pOpt[0]
rho_err = np.sqrt(pCov[0][0])


print(f"measured density = {rho * 10**3} +/- {rho_err * 10**3}g/ml")
dx = np.linspace(0,200,100)
ax1.plot(dx,f_mass(dx,rho)*1000)
ax1.set(xlabel = 'Volume [$\mu$M]', ylabel = 'Mass [mg]')
ax2.set(xlabel = 'Volume [$\mu$M]', ylabel = 'Residual Difference [mg]')
res = mass - rho* vol
ax2.plot(vol, res*1000,"C0.")
plt.axhline(0,ls = "--",color = "Grey")

plt.savefig("/work/notes/pipette_callibration.pdf",dpi=300)
```

```
line = np.polynomial.Polynomial.fit(dx,dx,1).convert()
print(line.coef)
plt.plot(dx, line(dx))
```
