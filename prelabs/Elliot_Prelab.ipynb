```
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

# 1

```
x = np.linspace(0,100,1000)
I = np.zeros(1000)
width = 10
I[100:100+width*10] = 1




def auto_correlate1(taus,I,x):
    ret = []
    for tau in taus:
        I_tau = np.copy(I)
        
        I_tau[:-tau*10] = I[tau*10:]
        I_tau[-tau*10:] = I[0]
        ret.append(1/np.max(x)*integrate.simpson(I_tau*I,x = x))
    return ret

fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize = (10,8))
fig.suptitle('autocorrelation of single pulse')
ax1.plot(x,I)
ax1.set(xlabel = 'time (ms)', ylabel = 'intensity')
print(integrate.simpson(I,x = x))

tau = np.arange(1,50, dtype = int)
G = auto_correlate1(tau,I,x)


ax2.plot(tau,G)
ax2.set(xlabel = '$\\tau$ (ms)', ylabel = 'G')

ax3.semilogy(tau,G)
ax2.set(xlabel = '$\\tau$ log(ms)', ylabel = 'G')
fig.tight_layout()
fig.savefig('elliot_prelab_Q1')
```

# 2

```
width = 10
up = 2*np.ones(width*10)
down = np.zeros(width*10)

pulse = np.concatenate((up,down))
x = np.linspace(0,100,len(pulse)*5)
print(len(x))
I = np.tile(pulse,5)
plt.plot(x,I)
```

```
plt.plot(x, I - 1)
```

```
def auto_correlate2(taus,I,x):
    ret = []
    for tau in taus:
        I_tau = np.copy(I)
        
        I_tau[:-tau*10] = I[tau*10:]
        I_tau[-tau*10:] = I[:tau*10]
        delta_I_tau = I_tau - 1
        delta_I = I - 1
        
        ret.append(1/np.max(x)*integrate.simpson(delta_I_tau*delta_I,x = x))
    return ret

fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize = (10,8))
fig.suptitle('autocorrelation of single pulse')
ax1.plot(x,I)
ax1.set(xlabel = 'time (ms)', ylabel = 'intensity')


tau = np.arange(1,50, dtype = int)
G = auto_correlate2(tau,I,x)


ax2.plot(tau,G)
ax2.set(xlabel = '$\\tau$ (ms)', ylabel = 'G')

ax3.semilogx(tau,G)
ax2.set(xlabel = '$\\tau$ log(ms)', ylabel = 'G')
fig.tight_layout()

fig.savefig('elliot_prelab_Q2')
```
